```python
import pandas as pd
df_pokemon = pd.read_csv('https://raw.githubusercontent.com/murpi/wilddata/master/pokemon.csv')
df_pokemon.head()
```

```python
df_pokemon.info()
```

```python
df_pokemon.isna().sum()
```

```python
df_pokemon.drop("Type 2", axis=1, inplace=True)
df_pokemon.info()
```

```python
df_pokemon_nonlegendary = df_pokemon.loc[df_pokemon["Legendary"] == False]
```

```python
df_pokemon_nonlegendary
```

```python
#selectionne les colonnes numériques
X = df_pokemon_nonlegendary\
              .select_dtypes('number')\
              .drop(columns = "#")
X
#X est basé sur les pokemons non légendaires
```

```python
from sklearn.neighbors import NearestNeighbors

modelNN = NearestNeighbors(n_neighbors=3)
modelNN.fit(X)
```

```python
df_pokemon.loc[df_pokemon['Name']\
               .str.contains('Giratina')]
```

```python
list_champion = ['Mewtwo','Lugia','Rayquaza','Giratina Origin Forme', 'Dialga', 'Palkia']
```

```python
df_pokemon.loc[df_pokemon['Name'] == 'Mewtwo', X.columns]
```

```python
neighbors = modelNN.kneighbors(df_pokemon.loc[df_pokemon['Name'] == 'Mewtwo', X.columns])
neighbors
```

```python
for Pokemon in list_champion:
    neighbors = modelNN.kneighbors(df_pokemon.loc[df_pokemon['Name'] == Pokemon, X.columns])

    print(f"Recommandation pour {Pokemon} :")

    pok_ind = neighbors[1][0]
    pok = df_pokemon_nonlegendary['Name'].iloc[pok_ind]
    #attention ici car le modèle a été entrainé sur X

    print("Pokemons proches : ", list(pok))
    print("Distances respectives : ", neighbors[0][0])
    print()
```

```python
for Pokemon in list_champion:
    type_1 = df_pokemon.loc[df_pokemon['Name'] == Pokemon]['Type 1'].values[0]

    same_type_pokemons = df_pokemon_nonlegendary.loc[df_pokemon_nonlegendary['Type 1'] == type_1]

    X = same_type_pokemons\
                  .select_dtypes('number')\
                  .drop(columns = "#")

    modelNN = NearestNeighbors(n_neighbors=3)
    modelNN.fit(X)


    neighbors = modelNN.kneighbors(df_pokemon.loc[df_pokemon['Name'] == Pokemon, X.columns])

    print(f"Recommandation pour {Pokemon} (Type 1 : {type_1}):")

    pok_ind = neighbors[1][0]
    pok = same_type_pokemons['Name'].iloc[pok_ind]
    #attention ici car le modèle a été entrainé sur X

    print("Pokemons proches : ", list(pok))
    print("Distances respectives : ", neighbors[0][0])
    print()


```
