You must correct each of the scripts below **using exception handling** and **without modifying lists**.

Script number 1

```python

list_integers = [1, 12, -45, 6, 27, 0, 24]
for i in list_integers:
    try:
        result = 1 / i
        print(result)
    except ZeroDivisionError:
      print(f'Can not divide by zero for integer {i}')
```

Script number 2

```python
list_integers = [1, 12, -45, 6, 27, 24]
for i in range(10):
    try:
        result = 1 / list_integers[i]
        print(result)
    except IndexError:
      print(f'Index {i} is out of range for the given list')
```

Script number 3

```python
list_integers = [1, 12, -45, 6, 27, 0, 24]
for i in range(10):
    try:
        result = 1 / list_integers[i]
        print(result)
    except IndexError:
      print(f'Index {i} is out of range for the given list')
    except ZeroDivisionError:
      print(f'Can not divide by zero for element {i}')
```

Script number 4

```python
list_words = ['hello', 'good morning', 'GOOD Evening', 34, 'Bye' ]

for word in list_words:
    try:
        print(str(word).upper())
    except AttributeError:
      print(f'Can not convert {type(word)} to uppercase')
```
