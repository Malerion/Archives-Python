# Function to automatic merge and fill missing values



## Create the function
You have to create a function which take 2 arguments :
- argument 1 : DataFrame main
- argument 2 : DataFrame opinion

Your function will return a new DataFrame which is the left merge of both DataFrames, and which has no missing values (so your function has to fill them), add new columns (flag, MONTH and DAY), and OPINION should be translated.

So you have to copy each useful code from the previous quests, and integrate it into a function. **This should achieve the same results as the previous quests.**

```
import pandas as pd
link_main = "https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather_main_2018.csv"
link_opinion = "https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather_opinion_2018.csv"
df_main = pd.read_csv(link_main)
df_opinion = pd.read_csv(link_opinion)
print(df_main.head())
print(df_opinion.head())
```

```
def merge_and_fill(DataFrameMain, DataFrameOpinion):
  # MERGE
    NewDataFrame = pd.merge(left=DataFrameMain,
                         right=DataFrameOpinion,
                         left_on='DATE',
                         right_on='date',
                         how='left')


  # FILL IN MISSING VALUES
    NewDataFrame['OPINION'].fillna(value='unknown', inplace=True)


  # ADD NEW COLUMNS
    NewDataFrame['flag'] = 1
    NewDataFrame['MONTH'] = pd.to_datetime(NewDataFrame['DATE']).dt.month
    NewDataFrame['DAY'] = pd.to_datetime(NewDataFrame['DATE']).dt.day

  # TRANSLATE OPINION
    def translate(opinion):
        translation_dict = {
            'very bad': 'très mauvais',
            'bad': 'mauvais',
            'not good not bad': 'ni bon ni mauvais',
            'good': 'bon',
            'very good': 'très bon'
        }
        return translation_dict.get(opinion, opinion)
    NewDataFrame['OPINION'] = NewDataFrame['OPINION'].apply(translate)


    return NewDataFrame

result = merge_and_fill(df_main, df_opinion)

result.head()
```

## Execution with a loop
You will execute your function in a 9 rounds loop, once per year, and agregate this in a new big DataFrame with 24 columns and lots of rows.

```
# Each year between 2010 and 2018 (included) has 2 sources : weather_main_YYYY.csv and weather_opinion_YYYY.csv

# Création d'un data frame vide pour agréger les résultats
df_total = pd.DataFrame()

for year in range(2010, 2018):
  #on construits les liens pour les dataframes
  link_main = f"https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather_main_{year}.csv"
  link_opinion = f"https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather_opinion_{year}.csv"

  # Lire les DataFrames
  df_main = pd.read_csv(link_main)
  df_opinion = pd.read_csv(link_opinion)

  #execution de la fonction merge and fill pour l'année en cours
  result = merge_and_fill(df_main, df_opinion)

  # concaténation du resultat actuelk avec le dataframe total
  df_total = pd.concat([df_total, result])

df_total.head(500)
```

## Add 2019 to 2022

```
for year in range(2019, 2023):
    link_main = "https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather{}.csv".format(str(year))
    df_main = pd.read_csv(link_main)
    df_total = pd.concat([df_total, df_main])
df_total
```


Find appropriate dataviz to answer to this question :

Which year had the best weather? Are years very different or not ?


```
# Convertir la colonne de date en type datetime
df_total['DATE'] = pd.to_datetime(df_total['DATE'])

# Calculer la moyenne annuelle du paramètre météorologique
annual_avg = df_total.groupby(df_total['DATE'].dt.year)['MAX_TEMPERATURE_C'].mean().reset_index()

# Créer le graphique en ligne
fig = px.line(annual_avg, x='DATE', y='MAX_TEMPERATURE_C', labels={'DATE': 'Année', 'MAX_TEMPERATURE_C': 'Température maximale moyenne'})
fig.update_layout(title='Moyenne annuelle de la température maximale', xaxis_title='Année', yaxis_title='Température maximale moyenne')
fig.show()

```

Le graphique en ligne serait une visualisation appropriée pour répondre à la question "Quelle année a eu le meilleur temps ?" Ceci permettrait une comparaison rapide des modèles météorologiques au fil des années.
