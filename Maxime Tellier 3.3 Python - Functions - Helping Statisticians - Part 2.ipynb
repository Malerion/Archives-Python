```
# Mission 12: Create a function that takes a number n as parameter, and then returns a list containing n lists, each containing n empty lists.
def generate_empty_lists(n):
    # Création d'une liste de n listes, chacune contenant n listes vides
    result = [[] for _ in range(n)]
    for inner_list in result:
        inner_list.extend([[] for _ in range(n)])
    return result

# Demander à l'utilisateur d'entrer un nombre pour n listes
user_input = int(input("Enter a number (n): "))

# Appeler la fonction avec l'entrée de l'utilisateur
result_lists = generate_empty_lists(user_input)

# Afficher le résultat
print("Result:")
for inner_list in result_lists:
    print(inner_list)
```
