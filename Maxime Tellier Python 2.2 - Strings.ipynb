Now, count only non-space characters and display how many there is.
Remember, using a for loop on a string will browse all the individual characters of it!

````
s = "Taumata whakatangihanga koauau o tamatea turi pukaka piki maungah oronuku pokai whenuaki tanatahu"
#Si "characters veut dire un mot c'est ce code"
# Creation de la variable pour le nombre de mots
word_count = 0

# Creation un indicateur pour vérifier si nous sommes à l'intérieur d'un mot
inside_word = False

# Parcour chaque caractère de la chaîne
for i in s:
    # Verification si le caractère est un espace (space)
    if i.isspace():
        # Si nous sommes à l'intérieur d'un mot, incrémentez le nombre de mots
        if inside_word:
            word_count += 1
        inside_word = False
    else:
        #Nous sommes à l'intérieur d'un mot
        inside_word = True

# Après la boucle, vérification si nous sommes à l'intérieur d'un mot à la fin de la chaîne
if inside_word:
    word_count += 1

print("count of non space characters is:", word_count)


````

# Mission 2

Ask the user for a string, then for a character (we assume user input follows the rules). Then, display the positions, **starting with 1**, of all occurrences of the character within the string.

For example, for the word "welcome" and for the character "e", the script will display:
```
position: 2
position: 7
```

````
user_string = input("Enter a string:")

search_char = input("Enter a character to search for:")

positions = []
for index, char in enumerate(user_string, start=1):
  if char == search_char:
    positions.append(index)
if positions:
  print(f"Character '{search_char}' was found at position:", positions)
else:
  print(f"Charcter '{search_char}' was not found")
````

# Mission 3

Ask the user for a string, then for an integer. Then, display the string repeated as many times as the given integer.

For example, `badger` and `4` will display `badgerbadgerbadgerbadger`

````
user_string = input("Enter a character: ")
repeat_count = int(input("Enter an integer: "))

repeated_string = ""
for _ in range(repeat_count):
    repeated_string += user_string

print("Result:", repeated_string)
````

# Mission 5

Ask the user for a string, then for a character, then for another character.

Then, display the string where the occurences of the first character are replaced by the second one.  
For example:
* string = `"Boulgour"`  
* 1st_char = `"g"`
* 2nd_char = `"v"`

Result : `"Boulvour"`

````
user_string = input("Enter a string: ")

first_char = input("Enter the first character to replace: ")
second_char = input("Enter the second character: ")

result_string = user_string.replace(first_char, second_char)
print("Result:", result_string)
````

# Mission 6

Ask the user for a string. Then, display the same string with all vowels converted to uppercase, and all consonants converted to lowercase.  
For example:
* string = `"Papa's got a brand new bag"`  
Result : `"pApA's gOt A brAnd nEw bAg"`


````
user_string = input("Enter a string: ")
result_string = ""

vowels = "AEIOUaeiou"

for char in user_string:
    if char in vowels:
        result_string += char.upper()
    else:
        result_string += char.lower()

print("Result:", result_string)
````
