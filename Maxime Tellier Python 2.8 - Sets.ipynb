There are four collection data types in the Python programming language:

* `List` is a collection which is ordered and changeable. Allows duplicate members.
* `Tuple` is a collection which is ordered and unchangeable. Allows duplicate members.
* `Set` is a collection which is unordered, unchangeable, and unindexed. No duplicate members, but you can remove items and add new items.
* `Dictionary` is a collection which is ordered and changeable. No duplicate members.

Question :    
If you display a `set` object, it's not display with the same order you define it. What can you say about the order the element are display ?

Answer : The order in which elements are displayed when we print a set object in Python is not guaranteed to be the same as the order in which you defined or added the elements. Sets are unordered collections, which means they do not maintain a specific order for their elements. This behavior is by design, and it allows sets to provide fast membership tests and eliminate duplicates efficiently. While the elements may sometimes appear in a particular order due to implementation details, you should not rely on this order, and it can change between different Python versions or implementations.


Question :    
Find a way to display the elements that exist in both sets.

```python
a_set = {"banana", "grapes", "cherry", "pineapple", "kiwi", "strawberry"}
b_set = {"lemon", "kiwi", "grapes", "blueberry", "watermelon"}

# Utiliser intersection_update pour modifier 'a_set' afin qu'il contienne uniquement les éléments présents dans les deux ensembles
a_set.intersection_update(b_set)

# Les éléments communs sont maintenant dans 'a_set'
elements_communs = a_set

# Afficher les éléments communs
print(elements_communs)
```
