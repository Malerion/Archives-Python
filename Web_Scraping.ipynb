```python
import requests

url = "http://www.chucknorrisfacts.fr/facts/top/1"

user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.6099.71 Safari/537.36'

response = requests.get(url, headers={'User-Agent': user_agent})

print(response.status_code)


```

```python
from bs4 import BeautifulSoup

html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
chuck_norris_blocks = soup.find_all('div', class_='card')

print(f"Il y a {len(chuck_norris_blocks)} blocs de blagues sur la page.")

if len(chuck_norris_blocks) >= 8:
    eighth_block_text = chuck_norris_blocks[7].text
    print(f"Texte du 8e bloc de blague : {eighth_block_text}")
```

```python
#sans les notes
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
chuck_norris_blocks = soup.find_all('div', class_='card')

print(f"Il y a {len(chuck_norris_blocks)} blocs de blagues sur la page.")

if len(chuck_norris_blocks) >= 8:
    eighth_block_text = chuck_norris_blocks[7].find('p', class_='card-text').text.strip()
    print(f"Texte de la 8e blague : {eighth_block_text}")
```

```python
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
chuck_norris_blocks = soup.find_all('div', class_='card')

print(f"Il y a {len(chuck_norris_blocks)} blocs de blagues sur la page.")

if len(chuck_norris_blocks) >= 8:
    eighth_block_note = chuck_norris_blocks[7].find('span', id='moyenne_100410').text.strip()
    print(f"Note de la 8e blague : {eighth_block_note}")
```

```python
blagues_dict = {}

for index, block in enumerate(chuck_norris_blocks, start=1):

    blague_text = block.find('p', class_='card-text').text.strip()


    footer_element = block.find('div', class_='card-footer')


    note_element = footer_element.find('span')


    blague_note = note_element.text.strip() if note_element else 'Note pas dispo'
    blagues_dict[f"Blague {index}"] = {'Blague': blague_text, 'Note': blague_note}

for key, value in blagues_dict.items():
    print(f"{key}: {value}")
```
