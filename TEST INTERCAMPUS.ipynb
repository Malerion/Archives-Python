Pour comprendre quand et comment utiliser les sets, prenons un exemple.

Tu es en train d'écrire un morceau de code. Tu es en train de taper avec certitude, quand soudain tu t'arrêtes, car tu viens d'arriver à un point du code où tu dois compter toutes les valeurs distinctes dans une liste et mettre en évidence celle qui est la plus utilisée. La raison pourrait être par exemple pour chercher à connaître le client le plus fréquent du magasin qui t'as embauché ou le film le plus regardé sur Netflix.

Pour les besoins de notre exemple, nous considérerons une liste de nombres:

```python
list_of_values = [24, 55, 19, 71, 42, 48, 37, 25, 74, 17, 65, 72, 91, 70, 47, 50, 15, 48, 27, 60, 67, 3, 20, 82, 57, 9, 76, 70, 89, 19, 93, 26, 85, 87, 0, 15, 80, 23, 78, 78, 33, 36, 31, 72, 57, 0, 40, 28, 56, 8, 62, 78, 95, 31, 51, 39, 0, 14, 73, 75, 60, 38, 97, 21, 43, 44, 17, 23, 79, 69, 10, 95, 73, 8, 30, 70, 43, 2, 7, 30, 68, 59, 79, 4, 50, 96, 19, 90, 23, 67, 15, 85, 27, 61, 10, 15, 62, 64, 30, 36, 11]
print(f'Our list is {len(list_of_values)} values long.')
```

Dans le domaine de la Data, nous devons souvent traiter un grand nombre de valeurs. Il n'est pas vraiment possible de les examiner toutes pour déterminer les valeurs distinctes à prendre en compte.
Donc, allons-y et utilisons des sets.

```python
set_of_values = set(list_of_values)
print(f'Our list is {len(set_of_values)} values long.')
```

Comme tu peux le voir, il y avait un certain nombre de valeurs uniques. Il aurait fallu un peu de temps pour les trouver toutes.
Nous pouvons maintenant passer à la construction du dictionnaire.

```python
dictionary_for_count = {} # Notice the dictionary has the same {} sets have
for value in set_of_values:
  dictionary_for_count[value] = list_of_values.count(value)
print(dictionary_for_count)

numero_le_plus_utilise = max(dictionary_for_count, key=dictionary_for_count.get)
print("le numero le plus utilise est: ",numero_le_plus_utilise)
```
