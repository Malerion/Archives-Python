Challenge.
Borner une saisie utilisateur
Dans un notebook/Google Colab, fais saisir à l'utilisateur deux valeurs entières, représentant les bornes d'un intervalle. Ces valeurs peuvent être saisies dans n'importe quel ordre !
Pour faire saisir une valeur à l'utilisateur, tu dois utiliser la fonction input().
Dans un premier temps, tu peux simplement tester la fonction input() dans un google collab pour voir ce qu'il se passe lorsque tu exécutes la cellule de code, après avoir écrit input() dans la cellule de code.
Fais saisir ensuite un troisième entier à l'utilisateur.
Ton programme doit afficher si la 3ème valeur saisie, est:

dans l'intervalle
En dessous de la borne inférieure
Au dessus de la borne supérieure

Première excercice:

```python

borne1 = int(input("Taper borne1 : "))
borne2 = int(input("Taper borne2 : "))

if borne1 > borne2:
    borne1, borne2 = borne2, borne1


entree = int(input("Taper le troisième entier : "))

if entree < borne1:
    resultat = f"plus petit que {borne1}, la borne inférieure"
elif entree > borne2:
    resultat = f"plus grand que {borne2}, la borne supérieure"
else:
    resultat = f"qui est dans l'intervalle"


print(f"Vous avez entré {entree}, {resultat}")
print(f"Résultat : {borne2}")

```

Deuxième exercice:

```python

borne1 = int(input("Taper la première borne: "))
borne2 = int(input("Taper la deuxième borne: "))


if borne1 > borne2:
    borne1, borne2 = borne2, borne1

entree = int(input("Taper le troisième entier: "))


if borne1 <= entree <= borne2:
    resultat = f"qui est dans l'intervalle"
else:
    resultat = f"qui n'est pas dans l'intervalle"


print(f"Vous avez entré {entree}, {resultat}")
print(f"Résultat : {entree}")
```

Troisième exercice:

```python

borne1 = int(input("Taper la première borne: "))
borne2 = int(input("Taper la deuxième born: "))

if borne1 > borne2:
    borne1, borne2 = borne2, borne1

entree = int(input("Saisissez le troisième entier : "))

if borne1 <= entree <= borne2:
    resultat = f"qui est dans l'intervalle"
elif entree < borne1:
    resultat = f"plus petit que {borne1}, la borne inférieure"
else:
    resultat = f"plus grand que {borne2}, la borne supérieure"

print(f"Vous avez entré {entree}, {resultat}")
print(f"Résultat : {entree}")
```
