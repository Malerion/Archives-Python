```
# Mission 3 : Create a function that uses the np.random.random() function to return an array with 10 random elements,
# all of which must be greater than 0.25 and less than 1.0

# Example of expected result: array([0.3897948 , 0.82850041, 0.5499838 , 0.81940137, 0.92175875, 0.89634373, 0.26821861, 0.72376496, 0.81098623, 0.28192741])

def random_array_generator():
    random_array = np.random.random(10)
    scale_array = 0.75 * random_array + 0.25
    return scale_array

result_array = random_array_generator()
print(result_array)
result_array
```

```
# Mission 4 : What does the np.empty() method do? Don't just copy the documentation, explain in your own words here

# Alors de ce que j'ai compris, np.empty() permet de créer un tableau non initialisé (c.a.d que les éléments du tableau ne sont pas définit par nous même
# lors de sa création).  Elle n'initialise pas les éléments du tableau à une valeur particulière, de sorte que le contenu du tableau peut être n'importe quelle
# donnée aléatoire qui existe déjà dans l'emplacement mémoire assigné au tableau.
# la syntaxe de np.empty() est comme ça:  numpy.empty(shape, dtype=float, order='C')
#c'est important à savoir que les éléménts créer dans ce tableau vide se sont pas garanties d'être zero ou un value spécifique contrairement à np.zeros() ou np.ones().
```
