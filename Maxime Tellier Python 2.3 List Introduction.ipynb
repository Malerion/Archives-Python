```python
names = ["Fredy", "Nicolas", "Alexandre"]

for i in enumerate(names):
  print(i,type(i))
  #La fonction enumerate() est une fonction intégrée qui est utilisée pour énumérer les éléments d'une séquence (comme une liste, un tuple ou une chaîne de caractères) tout en fournissant à la fois l'indice de chaque élément et la valeur de l'élément correspondant. Elle renvoie un objet énumérable qui peut être itéré à l'aide d'une boucle for.
  #Grâce au type(i) on peut constater que les objets son retournée comme tuple à leur sortie.
```

1. What type of object is returned as output?  
Hint: You can use `type(i)`.

Answer: La fonction enumerate() est une fonction intégrée qui est utilisée pour énumérer les éléments d'une séquence (comme une liste, un tuple ou une chaîne de caractères) tout en fournissant à la fois l'indice de chaque élément et la valeur de l'élément correspondant. Elle renvoie un objet énumérable qui peut être itéré à l'aide d'une boucle for.
  
  Grâce au type(i) on peut constater que les objets son retournée comme tuple à leur sortie.


```python
for i, val in enumerate(names):
  print(i)
```

```python
for i, val in enumerate(names):
  print(val)
```

2. What does `i` correspond to?  
What does `val` correspond to?  
Why is it so?

Answer: La variable i correspond à l'indice de chaque élément dans la liste names. Lorsqu'on utilise enumerate() dans cette boucle, elle renvoie des paires de (indice, valeur) pour chaque élément dans l'itérable names. Ainsi, i représente l'indice de chaque élément dans la liste, en commençant par 0 et en s'incrémentant de 1 à chaque itération de la boucle.

La variable val correspond à la valeur de chaque élément dans la liste names. Lorsqu'on utilise enumerate() dans la boucle, elle renvoie des paires de (indice, valeur) pour chaque élément dans l'itérable names. Donc, val représente la valeur réelle de chaque élément dans la liste au fur et à mesure que la boucle itère à travers elle.

Cette distinction se produit parce que dans la première boucle (for i, val in enumerate(names):), il est assigné l'indice à la variable i et la valeur à la variable val. Dans la deuxième boucle (for i, val in enumerate(names):), ça été inversé les position, de sorte que val reçoit maintenant la valeur (l'élément), et i reçoit l'indice. L'ordre des variables dans l'assignation détermine ce qu'elles représentent à chaque itération de la boucle.

```python
for val, i in enumerate(names):
  print(i)
```

```python
for val, i in enumerate(names):
  print(val)
```

3. This time, `i` returns first names and `val` returns numbers. Why is it so?

Answer: Ici, val reçoit l'indice (entier) de chaque élément, et i reçoit la valeur (élément) de chaque élément. du coup, la boucle affiche les noms (valeurs).




4. What does the built in function `enumerate` do?
In what cases can it be useful?

Answer: La fonction enumerate est utilisée pour itérer sur les éléments d'un itérable comme une liste, un tuple, une chaîne de caractères tout en gardant une trace de l'indice de chaque élément. Elle renvoie des paires de (indice, élément) où l'indice est la position de l'élément dans l'itérable, et l'élément est la valeur à cette position. ça peut servir pour avoir besoin d'accéder à la fois à l'indice et à la valeur des éléments dans un itérable au sein d'une boucle.
Ca peut servir pour Itérer sur des listes et je pense créer des dictionnaires.

3. What does `range(2,8,2)` mean ?  
Why are only the elements 'name', 'hi' and 'name' displayed?

Answer: La fonction range(2, 8, 2) signifie :

-Le premier argument, 2, indique le point de départ de la séquence. La séquence commencera à partir de l'indice 2 inclus.
-Le deuxième argument, 8, indique la fin de la séquence. La séquence ira jusqu'à l'indice 8 exclus.
-Le troisième argument, 2, indique le pas entre chaque valeur de la séquence. Dans ce cas, la séquence avancera de 2 en 2.
La boucle for utilise cette séquence générée par range(2, 8, 2) pour itérer sur les indices de 2 à 8 (non inclus), avec un pas de 2.

C'est pourquoi seuls les éléments 'name', 'hi' et 'name' sont affichés. La séquence générée parcourt les indices 2, 4 et 6, correspondant aux éléments 'name', 'hi' et 'name' dans la liste
