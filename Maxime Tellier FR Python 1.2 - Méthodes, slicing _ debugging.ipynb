3.6 Cette fois, on veut connaître les variables dont les valeurs ont un quotient supérieur à 5 lorsqu'on divise chaque valeur par 5. Remplacez "..." par le code qui convient.

```python
var_1 = 34
var_2 = 37
var_3 = 24
var_4 = 29
var_5 = 31

if var_1 // 5 > 5:
  print(f"Le quotient de var_1 divisé par 5 est égal à {var_1 // 5}.")
if var_2 // 5 > 5:
  print(f"Le quotient de var_2 divisé par 5 est égal à {var_2 // 5}.")
if var_3 // 5 > 5:
  print(f"Le quotient de var_3 divisé par 5 est égal à {var_3 // 5}.")
if var_4 // 5 > 5:
  print(f"Le quotient de var_4 divisé par 5 est égal à {var_4 // 5}.")
if var_5 // 5 > 5:
  print(f"Le quotient de var_5 divisé par 5 est égal à {var_5 // 5}.")
```
