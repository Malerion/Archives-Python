```python
import json
import requests
import pandas as pd

link = "https://raw.githubusercontent.com/WildCodeSchool/java-jsonparse/master/weather.json"

# This function gets the file content from the URL
r = requests.get(link)

# You can access to the text content with this function
r.text

data = json.loads( r.text)
data
```

```python
city_name = data['name']
print(city_name)
```

```python
latitude = data['coord']['lat']
longitude = data['coord']['lon']
print('City latitude:' ,{latitude})
print('City longitude:',{longitude})
```

```python
for item in data['weather']:
    main_content = item['main']
    print("Weather:", main_content)
```

```python
# You can directly read JSON from your request object
print(r.json())

# And you can import JSON object to pandas
pd.json_normalize(r.json())
```

```python
pd.json_normalize(r.json(),
                  record_path= "weather",
                  meta = 'dt')
```

```python

df = pd.json_normalize(r.json(), record_path=['weather'], meta=['name', 'visibility'])


selected_columns = ['main', 'name', 'visibility']
filtered_df = df.loc[:, selected_columns]

filtered_df


```
