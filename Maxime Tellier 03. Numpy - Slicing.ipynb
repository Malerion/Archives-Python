# Mission 2

Modify, using a `numpy` method, the `array_1` object to be an object that has two axes, including 1 row and 32 columns.  
Save it under the variable `array_2`.

```python
array_2 = np.reshape(array_1, (1, 32))
print(array_1)
print(array_2)
```

Now, modify `array_1` into an object that has 32 rows and only 1 column.  
Save it under the variable `array_3`.

```python
array_3 = np.reshape(array_1, (32, 1))
print(array_1)
print(array_3)
```

Access the first element of `array_2`, then of `array_3`.  
What do you notice?  
Why ?

```python
first_element_a2 = array_2[0, 0]
first_element_a3 = array_3[0, 0]

print(first_element_a2)
print(first_element_a3)
# On remarque que c'est le même élément que dans array_1 ce qui veut dire que le reshape ne modifie pas les ordres des éléments.
# Cependant on remarque que contrairement à l'excercie précédent quand on cherchait le premier element de array_1 nous avions juste besoin de [0]
# Alors que pour array_2 et 3 nous devons faire [0,0] car la notation d'indices et différente entre tableau unidimensionel et multidimensionel
```

# Mission 4

Reshape `array_1` to be a 4-row, 8-column array,
rather than a 32-element vector.  
Save your array under the variable `array_4`.

```python
array_4 = np.reshape(array_1, (4, 8))
print("Original array_1:")
print(array_1)
print("\nReshaped array_4:")
print(array_4)





```

Access for each row, the values ​​of the columns which have the index 1 to 2 included.

```python
array_4 = np.array([[10, 40, 45, 5, 66, 91, 43, 75],
                    [40, 28, 55, 45, 31, 71, 67, 54],
                    [54, 35, 64, 27, 12, 93, 17, 91],
                    [50, 57, 74, 61, 86, 20, 36, 35]])
values_columns_1_to_2 = array_4[:, 1:3]
print("Values for each row from columns with indices 1 to 2:")
print(values_columns_1_to_2)
```

Access for rows at position 1 and 2 inclusive, the values ​​of all columns.

```python
values_rows_1_to_2 = array_4[1:3, :]
print("Values for rows at positions 1 and 2 for all columns:")
print(values_rows_1_to_2)
```

Access for the rows which have the index 2 to 3 inclusive, to the values ​​of the columns which have the index 3 to 5 inclusive.

```python
values_rows_2_to_3_columns_3_to_5 = array_4[2:4, 3:6]
print("Values for rows with indices 2 to 3 and columns with indices 3 to 5:")
print(values_rows_2_to_3_columns_3_to_5)
```

Select the first column, as well as the columns whose index is a multiple of 2.

```python
selected_columns = array_4[:, ::2]
print(selected_columns)
```

Select columns whose index is odd.

```python
selected_columns_odd = array_4[:, 1::2]
print(selected_columns_odd)
```
