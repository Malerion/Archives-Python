```
link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)
# on va exploser la colonne actors_list
df_exploded = df.explode('actors_list')

#on utilise la lambada pour retirer les u' de actors list
df_exploded['actors_list'] = df_exploded['actors_list'].apply(lambda x: [actor.replace("u'", "'") for actor in eval(x)])

#utilisation de value_count pour compter les occurences de chaque acteurs
actor_counts = df_exploded['actors_list'].value_counts()

#on prÃ©pare le rÃ©sultat
most_mentioned_actor = actor_counts.idxmax()
print(f"The actor mentioned most often is: {most_mentioned_actor}")
```
