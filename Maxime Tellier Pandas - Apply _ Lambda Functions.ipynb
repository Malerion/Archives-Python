### Welcome to the Apply Challenge

```python
link = 'https://raw.githubusercontent.com/chriszapp/datasets/main/books.csv'
books = pd.read_csv(link, nrows = 13)
books.head()
```

```python
# Firstly, have a look at the dataset

```

```python
# Mission 1: use apply to create a new column called 'reviews_opinion': if the ratings is between 4.5 and 5, the reviews_opinion is 'very good';
# if the rating is between 4 and 4.5, the reviews_opinion is 'good'; if the rating is between 3 and 4, the reviews_opinions will be 'neutral';
# if the rating is up to 3 the reviews_opinion is going to be 'bad'

# Add your code here
books['reviews_opinion'] = books['average_rating'].apply(lambda x: 'very goof' if 4.5 <= x <= 5 else ('good' if 4 <= x < 4.5 else ('neutral' if 3 <= x < 4 else 'bad')))
books.head()
```

```python
# Mission 2: using apply with a function, create a column with the number of authors per book
# (note that in the column 'authors', each name of the author is divided with the character '/')

# Add your code here
def count_authors(authors):
    return len(authors.split('/'))

books['num_authors'] = books['authors'].apply(count_authors)
books.head()
```

```python
# Mission 3: rewrite the same task as Mission 2 using the lamdba function

# Add your code here
books['num_authors'] = books['authors'].apply(lambda x: len(x.split('/')))

books.head()
```

```python
# Mission 4: create a column 'comment_ration'.
# Use the columns 'ratings_count' and 'text_reviews_count' to calculate the percentage of how many text review per reviews are per book:
# percentage = text / reviews * 100
# Use the mathematical operations between columns

# Add your code here
books['comment_ration'] = (books['text_reviews_count'] / books['ratings_count']) * 100
books.head()
```

```python
# Mission 5: obtain the same result of Mission 4 using apply with a function.
# Try both methods to apply to the whole dataset and apply to selected columns

# Add your code here
def calculate_comment_ratio(row):
    return (row['text_reviews_count'] / row['ratings_count']) * 100

books['comment_ration_apply_all'] = books.apply(calculate_comment_ratio, axis=1)

selected_columns = ['ratings_count', 'text_reviews_count']
books['comment_ration_apply_selected'] = books[selected_columns].apply(calculate_comment_ratio, axis=1)

books.head()
```

```python
# Mission 6: obtain the same result of Mission 5 using apply lambda
# Try both methods to apply to the whole dataset and apply to selected columns

# Add your code here
books['comment_ratio_lambda_all'] = books.apply(lambda row: (row['text_reviews_count'] / row['ratings_count']) * 100, axis=1)

selected_columns = ['ratings_count', 'text_reviews_count']
books['comment_ratio_lambda_selected'] = books[selected_columns].apply(lambda row: (row[1] / row[0]) * 100, axis=1)

books.head()
```

```python
# Mission 7: extract the year from the publication_date column. How many years of publications are recorded in this dataset? Which are them?

books['publication_date'] = pd.to_datetime(books['publication_date'], errors='coerce')

books['publication_year'] = books['publication_date'].dt.year


unique_years = books['publication_year'].unique()


num_years = len(unique_years)

print("Unique Years of Publications:", unique_years)
print("Number of Years Recorded:", num_years)

```
