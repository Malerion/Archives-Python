# Mission 5

Rename the columns [of this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv), try using the following 2 different methods to achieve this:
- Setting column names directly when reading the file
- Editing after creating the dataframe ([This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html))

```
import pandas as pd
# Définir les noms de colonnes
column_names = ["Ville", "Couleur_decalred", "forme_declared", "Region", "Date"]

# Lire le fichier CSV avec les noms de colonnes spécifiés
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv", header=0, names=column_names)

df.head(5)

```

```
import pandas as pd

# Lire le fichier CSV avec la première ligne comme en-tête
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv", header=0)

# Définir les noms de colonnes
column_names = ["Ville", "Couleur_Reported", "Forme_Reported", "Region", "Date"]

# Renommer les colonnes avec .rename()
df.rename(columns=dict(zip(df.columns, column_names)), inplace=True)

# Afficher les 5 premières lignes de la dataframe
df.head(5)
```

```
import pandas as pd

# Lire le fichier CSV avec la première ligne comme en-tête
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv", header=0)

# Définir les noms de colonnes
column_names = ["Ville", "Couleur_Reported", "Forme_Reported", "Region", "Date"]

# Renommer les colonnes
df.columns = column_names

# Afficher les 5 premières lignes de la dataframe
df.head(5)

```
