Create a function that returns the sum of two numbers.  
The user of the function must be able to sum two numbers of his choice, when the function is executed.  
You must not use `input`.  


```python
def add_numbers(num1,num2):
  result = num1 + num2
  return result

#pour donner Ã  la fonction n'importe quel chiffre deux fois. l'utilisateur entre deux chiffres dans les ()
add_numbers(12,13)
```

Question:   
What is a parameter in a function?

Answer:  a parameter in a function is a variable that is used to receive input data or arguments when the function is called. Parameters act as placeholders or containers within the function and allow you to pass values into the function from the outside. These values can then be used and manipulated within the function's code.

Create a function that returns 2 objects when executed:
* An object that will be the product of two numbers chosen by the user of the function.
* A word chosen by the user of the function.

```python
def exec_objs(num1,num2,word):

  product = num1 * num2
  return [product, word]

exec_objs(5,10,"Space Marines")
```

# Mission 5

Create a function that returns the sum of two numbers when executed.  
Also, your function must return a sentence that simply indicates if the sum of the two numbers chosen by the user is:
* less than 10
* between 10 and 50
* greater than 50  
Again, you must use `return`.

```python
def sum_and_range(num1,num2):
  result = num1 + num2

  if result <10:
    return"The sum of {} and {} is {}. It is less than 10.".format(num1,num2,result)
  elif 10 <= result <= 50:
    return"The sum of {} and {} is {}. It is between 10 and 50.".format(num1,num2,result)
  else:
    return"The sum of {} and {} is {. It is greater than 50}".format(num1,num2,result)

sum_and_range(10,30)
```
