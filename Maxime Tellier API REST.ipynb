```python
import requests
import pandas as pd

# URL de l'API
url = 'https://data.nantesmetropole.fr/api/records/1.0/search/?dataset=244400404_stations-velos-libre-service-nantes-metropole-disponibilites&q=&facet=banking&facet=bonus&facet=status&facet=contract_name'

# Effectuer une requête GET
response = requests.get(url)

# Vérifier si la requête a réussi (code de statut 200)
if response.status_code == 200:
    # Convertir les données JSON en un dictionnaire Python
    data = response.json()

    # Extraire la liste de records
    records = data.get('records', [])

    # Transformer la liste de records en DataFrame
    df = pd.json_normalize(records)

    # Sélectionner les colonnes nécessaires
    df_selected = df[["fields.address", "fields.available_bikes", "fields.bike_stands"]]

    # Calculer le taux de remplissage
    df_selected["filling_rate"] = df_selected["fields.available_bikes"] / df_selected["fields.bike_stands"]

    # Renommer les colonnes
    df_selected = df_selected.rename(columns={"fields.address": "adresse", "filling_rate": "taux_remplissage"})

    # Afficher le DataFrame (les 5 premières lignes)
    print(df_selected.head())
else:
    # Afficher un message d'erreur si la requête a échoué
    print(f"Erreur de requête. Code de statut : {response.status_code}")

```
