```python
x = 5
x_float = float(x)

# Pour vérifiez si la conversion de x_float a marché
if type(x_float) == float:
    print("La conversion en float a réussi.")
else:
    print("La conversion en float a échoué.")

```

```python
x = 5.4
x_int = int(x)

# Pour vérifier si la conversion de x_int à marché
if type(x_int) == int:
  print("La conversion en int a réussi.")
else:
  print("La conversion en int a échoué.")
```

# Built-in Function

3.1 Utilisez une [built-in function](https://docs.python.org/3/library/functions.html) pour compter le nombre de lettres dans l'objet de type `string` suivant:

```python
x = "Hello World"
lettres = [caractere for caractere in x if caractere.isalpha()]
nombre_de_lettres = len(lettres)

print("Le nombre de lettres dans la chaîne est :", nombre_de_lettres)

```

3.2 Utilisez une built-in function pour compter le nombre d'éléments dans l'objet de type `list` suivant:

```python
x = ["non", "pas", 1, "ni", 2, "ni", 3, "mais", 7.2132]
nombre_d_elements = len(x)

print("Le nombre d'éléments dans la liste est:", nombre_d_elements)
```

3.3 Utilisez une built-in function sur une des deux variables pour que `resultat`  contienne une valeur positive.

```python
var_1 = 4
var_2 = -5

resultat = abs(var_1) + abs(var_2)

print("Le résultat avec des valeurs absolues est :", resultat)
```

3.4 Utilisez une built-in function pour renvoyer la valeur minimum de la `list` suivante:

```python
a_list = [3.2, 41, 2, 23, 7]
valeur_minimale = min(a_list)
print("La valeur minimale dans cette liste est:", valeur_minimale)
```

3.5 Utilisez une built-in function pour renvoyer la somme des valeurs de la `list` suivante:

```python
a_list = [3.2, 41, 2, 23, 7]
somme_des_valeurs = sum(a_list)

print("La somme des valeurs de cette liste est:", somme_des_valeurs)
```
