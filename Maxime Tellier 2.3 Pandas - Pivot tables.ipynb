# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
# Your code here
pivot_table = df_weather.pivot_table(values=['MAX_TEMPERATURE_C', 'SUNHOUR'], index='MONTH', aggfunc='mean')
pivot_table = pivot_table.reset_index()

fig, ax1 = plt.subplots()

# Bar plot pour Temperature bleu
ax1.bar(pivot_table['MONTH'], pivot_table['MAX_TEMPERATURE_C'], color='blue', label='Temperature')
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (°C)', color='blue')
ax1.tick_params('y', colors='blue')

#deuxième axe  y pour Sunhours (rrouge)
ax2 = ax1.twinx()
ax2.plot(pivot_table['MONTH'], pivot_table['SUNHOUR'], color='red', label='Sunhours')
ax2.set_ylabel('Sunhours', color='red')
ax2.tick_params('y', colors='red')

plt.title('Average Temperature and Sunhours by Month')
```
