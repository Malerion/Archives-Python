# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```
def afficher_taux_remplissage(dataframe):
    """
    Affiche le taux de remplissage de chaque colonne du DataFrame.

    Parameters:
    - dataframe : pandas DataFrame

    Returns:
    None
    """
    # Calculer le taux de remplissage pour chaque colonne
    taux_remplissage = (dataframe.notnull().sum() / len(dataframe)) * 100

    # Afficher le résultat pour chaque colonne
    for colonne, taux in taux_remplissage.iteritems():
        print(f"{colonne}: {taux:.2f}%")


afficher_taux_remplissage(df)
```

# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```
taux_remplissage = df.count() / len(df) * 100
# Créer les catégories en utilisant cut
categories = pd.cut(taux_remplissage, bins=[0, 25, 50, 75, 100], labels=['0-25%', '25-50%', '50-75%', '75-100%'])

# Compter le nombre de colonnes dans chaque catégorie
nombre_colonnes_par_quartile = categories.value_counts()

# Afficher le résultat
print(nombre_colonnes_par_quartile)
```

# Mission 6

Un dataset trop pauvre en data n'est pas vraiment intéressant. Après avoir analyser les colonnes, nous nous sommes rendu compte que nous pouvons supprimer les colonnes avec moins de 60% de valeurs non nulles, sauf les colonnes pnns_groups_1 et pnns_groups_2, nutrition-score-fr_100g et nutrition-score-uk_100g. En effet ces 2 colonnes vont nous être utiles par la suite.

Si ce n'est pas déjà fait, supprime également les colonnes suivantes car elles ne sont pas vraiment utiles pour répondre à notre problématique :
* creator
* created_t
* created_datetime
* last_modified_t
* brands_tags
* countries
* countries_tags
* additives_tags
* states
* states_tags

```
# Liste des colonnes à conserver
colonnes_utiles = ['pnns_groups_1', 'pnns_groups_2', 'nutrition-score-fr_100g', 'nutrition-score-uk_100g']

# Sélectionner les colonnes avec moins de 60% de valeurs non nulles
colonnes_a_supprimer = df.columns[df.isnull().mean() > 0.4]
colonnes_a_supprimer = colonnes_a_supprimer[~colonnes_a_supprimer.isin(colonnes_utiles)]

# Supprimer les colonnes inutiles
df = df.drop(columns=colonnes_a_supprimer, errors='ignore')

# Supprimer les colonnes spécifiques
colonnes_a_supprimer_manuelles = ['creator', 'created_t', 'created_datetime', 'last_modified_t', 'brands_tags', 'countries', 'countries_tags', 'additives_tags', 'states', 'states_tags']
df = df.drop(columns=colonnes_a_supprimer_manuelles, errors='ignore')

```
